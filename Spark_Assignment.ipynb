```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
spark = SparkSession.builder \
    .appName("Spark Assignment") \
    .getOrCreate()
```

```python
df = spark.read.csv("employees.csv", header=True, inferSchema=True)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
selected_df = df.select("name", "salary", "department")
selected_df.show()
```

```python
filtered_df = df.filter(col("salary") > 50000)

filtered_df.show()
```

```python
df = df.withColumnRenamed("salary", "employee_salary")

df.show()
```

```python
from pyspark.sql.functions import col

df = df.withColumn("age", col("age").cast("int"))

df.printSchema()
```

```python
df = df.withColumn(
    "annual_salary",
    col("employee_salary") * 12
)

df.show()
```

```python
df.show()
```

```python
df = df.na.fill({
    "employee_salary": 0,
    "department": "Unknown"
})

df.show()
```

```python
group_df = df.groupBy("department").avg("employee_salary")
group_df.show()
```

```python
filtered_df = df.select(
    "id",
    "name",
    "department",
    "employee_salary"
)

filtered_df.show(5)
```

```python
high_salary = df.filter(df.employee_salary > 50000)
high_salary.show()
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
df.select(
    "name",
    "employee_salary",
    "annual_salary"
).show(5)
```

```python
from pyspark.sql.functions import col, when, count

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
dept_salary = df.groupBy("department") \
                .avg("employee_salary")

dept_salary.show()
```

```python
lazy_df = df.filter(df.employee_salary > 30000) \
            .select(
                "name",
                "department",
                "employee_salary"
            )
```

```python
lazy_df.show()
```

```python
lazy_df.explain(True)
```

CSV vs Parquet

CSV:
- Row-based storage
- Larger file size
- Slower queries
- No compression by default

Parquet:
- Columnar storage
- Compressed
- Faster analytics
- Supports Predicate Pushdown

# Spark Architecture

Spark follows a master-worker architecture.

## Components

### Driver Program
- Creates the SparkSession.
- Builds the DAG (Directed Acyclic Graph).
- Schedules and coordinates tasks.

### Cluster Manager
- Allocates CPU and memory resources to executors.

### Executors
- Execute tasks assigned by the Driver.
- Store intermediate data.
- Return results to the Driver.

### Execution Flow
Driver Program → Cluster Manager → Executors → Result

# Performance Insights

- Spark uses Lazy Evaluation.
- Transformations are executed only when an Action (show(), count(), collect()) is called.
- Filtering data early reduces processing time.
- GroupBy operations involve Shuffle, making them more expensive.
- Parquet provides better performance than CSV because it is a columnar storage format.
- Predicate Pushdown improves query performance by reading only the required data.

# Best Practices

- Read only the required columns.
- Use inferSchema or define an explicit schema.
- Filter data as early as possible.
- Avoid collect() for large datasets.
- Prefer show() for previewing data.
- Handle null values before analysis.
- Use Parquet instead of CSV for analytical workloads.
- Minimize shuffle operations for better performance.
